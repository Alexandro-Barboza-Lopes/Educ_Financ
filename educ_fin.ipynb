```python
import os
from google import genai
from google.generativeai import types
from IPython import get_ipython
from IPython.display import display
import os
from google.colab import userdata
import google.generativeai as genai
from datetime import datetime
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib.units import inch, cm
from reportlab.lib.enums import TA_CENTER, TA_JUSTIFY
import tempfile

# Configuração da API - substitua pela sua chave
# Você pode usar os.environ['GOOGLE_API_KEY'] = "sua_chave_aqui" diretamente no código
# Ou usar a abordagem do userdata do Google Colab como no exemplo original

userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))
# Inicializa o modelo e a sessão de chat
model = genai.GenerativeModel('gemini-2.0-flash')

# Definição das instruções para o comportamento do assistente
instrucoes_sistema = """
Você é um educador financeiro experiente, calmo e acolhedor com mais de 20 anos de experiência.
Seu papel é:
- Fornecer orientação financeira personalizada com base nas respostas do usuário
- Ser atencioso, empático e nunca julgar a situação financeira do usuário
- Oferecer conselhos práticos e realistas considerando a situação específica
- Enfatizar a importância de ter uma reserva de emergência de 3 a 6 vezes o valor total das despesas mensais
- Sugerir estratégias para gerenciar dívidas e melhorar a saúde financeira
- Manter um tom calmo, positivo e encorajador em todas as interações
- Compartilhar, quando apropriado, as estratégias de investimento de Luiz Barsi, especialmente sobre dividendos
"""

# Estratégias de investimento de Luiz Barsi - para uso nas orientações
estrategias_barsi = """
Estratégias de Investimento de Luiz Barsi:

1. Foco em empresas pagadoras de bons dividendos:
   - Buscar empresas sólidas com histórico de pagamento consistente de dividendos
   - Preferir empresas que distribuem regularmente pelo menos 6% ao ano em dividendos
   - Reinvestir os dividendos para criar o "efeito bola de neve"

2. Critérios de seleção de empresas segundo Barsi:
   - Empresas com forte governança corporativa
   - Líderes em seus segmentos
   - Baixo endividamento
   - Negócios simples de entender e pouco afetados por inovações tecnológicas
   - Histórico de resultados crescentes ou estáveis
   - Que atuam em setores essenciais da economia

3. Filosofia de longo prazo:
   - Investir pensando em décadas, não em dias ou meses
   - Não se preocupar com oscilações de curto prazo do mercado
   - Buscar uma renda passiva crescente através dos dividendos
   - Tornar-se "sócio" das empresas e não apenas especulador

4. Gestão da carteira:
   - Diversificar em 8-12 empresas de qualidade
   - Acompanhar os resultados trimestrais das empresas
   - Aumentar posição em empresas que mantêm bons fundamentos
   - Avaliar constantemente se os motivos que o levaram a investir na empresa continuam válidos

5. Recomendações práticas:
   - Começar com qualquer valor, mesmo que pequeno
   - Investir com regularidade (mensalmente)
   - Reinvestir os dividendos para acelerar o crescimento do patrimônio
   - Ter paciência e disciplina para colher os resultados no longo prazo
"""

# Cria um objeto de chat com as instruções do sistema
chat = model.start_chat(history=[])

# Função para obter respostas do modelo AI
def obter_resposta(mensagem, incluir_barsi=False):
    # Adiciona as instruções ao sistema para cada mensagem
    if incluir_barsi:
        resposta = chat.send_message(f"{instrucoes_sistema}\n\n{estrategias_barsi}\n\nUsuário: {mensagem}")
    else:
        resposta = chat.send_message(f"{instrucoes_sistema}\n\nUsuário: {mensagem}")
    return resposta.text

# Função para converter string em valor numérico, removendo caracteres não numéricos
def converter_para_numero(valor_str):
    # Remove R$, pontos, vírgulas e espaços, e substitui vírgula por ponto
    if isinstance(valor_str, str):
        valor_limpo = valor_str.replace("R$", "").replace(".", "").replace(",", ".").strip()
        # Tenta converter para float
        try:
            return float(valor_limpo)
        except ValueError:
            return 0
    elif isinstance(valor_str, (int, float)):
        return float(valor_str)
    else:
        return 0

# Função para formatar valores monetários
def formatar_moeda(valor):
    return f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

# Função para gerar relatório em PDF
def gerar_relatorio_pdf(dados_usuario):
    # Cria um arquivo temporário para o PDF
    pdf_filename = f"relatorio_financeiro_{datetime.now().strftime('%Y%m%d_%H%M%S')}.pdf"

    # Extrai e converte valores financeiros
    salario = converter_para_numero(dados_usuario.get('salario', 0))
    renda_extra = converter_para_numero(dados_usuario.get('renda_extra', 0))
    gastos = converter_para_numero(dados_usuario.get('gastos', 0))
    renda_total = salario + renda_extra

    # Cálculo da reserva de emergência recomendada
    reserva_3x = gastos * 3
    reserva_6x = gastos * 6

    # Estilos para o PDF
    styles = getSampleStyleSheet()

    # Estilo para o título principal
    title_style = ParagraphStyle(
        'TitleStyle',
        parent=styles['Heading1'],
        fontSize=18,
        alignment=TA_CENTER,
        spaceAfter=16,
        textColor=colors.darkblue,
    )

    # Estilo para os subtítulos
    subtitle_style = ParagraphStyle(
        'SubtitleStyle',
        parent=styles['Heading2'],
        fontSize=14,
        spaceAfter=10,
        textColor=colors.darkblue,
    )

    # Estilo para texto normal
    normal_style = ParagraphStyle(
        'NormalStyle',
        parent=styles['Normal'],
        fontSize=10,
        alignment=TA_JUSTIFY,
        spaceAfter=8,
    )

    # Estilo para informações financeiras
    info_style = ParagraphStyle(
        'InfoStyle',
        parent=styles['Normal'],
        fontSize=10,
        spaceAfter=6,
    )

    # Cria o documento PDF
    doc = SimpleDocTemplate(
        pdf_filename,
        pagesize=A4,
        rightMargin=72,
        leftMargin=72,
        topMargin=72,
        bottomMargin=72
    )

    # Lista de elementos para o PDF
    elements = []

    # Data do relatório
    data_atual = datetime.now().strftime("%d/%m/%Y")

    # Adiciona o título
    elements.append(Paragraph("Relatório de Planejamento Financeiro", title_style))
    elements.append(Paragraph(f"Data: {data_atual}", info_style))
    elements.append(Spacer(1, 0.5*cm))

    # Seção 1: Perfil Financeiro
    elements.append(Paragraph("1. Perfil Financeiro", subtitle_style))
    elements.append(Paragraph(f"<b>Situação atual:</b> {dados_usuario.get('pergunta_1', 'Não informado')}", normal_style))
    elements.append(Paragraph(f"<b>Estratégias atuais:</b> {dados_usuario.get('pergunta_2', 'Não informado')}", normal_style))
    elements.append(Paragraph(f"<b>Reserva de emergência:</b> {dados_usuario.get('pergunta_3', 'Não informado')}", normal_style))
    elements.append(Paragraph(f"<b>Objetivos financeiros:</b> {dados_usuario.get('pergunta_4', 'Não informado')}", normal_style))
    elements.append(Paragraph(f"<b>Interesse em investimentos:</b> {dados_usuario.get('pergunta_5', 'Não informado')}", normal_style))
    elements.append(Spacer(1, 0.5*cm))

    # Seção 2: Análise de Renda e Despesas
    elements.append(Paragraph("2. Análise de Renda e Despesas", subtitle_style))

    # Tabela de renda e despesas
    dados_tabela = [
        ["Item", "Valor"],
        ["Salário mensal", formatar_moeda(salario)],
        ["Renda extra mensal", formatar_moeda(renda_extra)],
        ["Renda total mensal", formatar_moeda(renda_total)],
        ["Gastos mensais", formatar_moeda(gastos)],
        ["Saldo mensal", formatar_moeda(renda_total - gastos)]
    ]

    t = Table(dados_tabela, colWidths=[doc.width/2.0]*2)
    t.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (1, 0), colors.lightblue),
        ('TEXTCOLOR', (0, 0), (1, 0), colors.white),
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 10),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('GRID', (0, 0), (-1, -1), 1, colors.lightgrey),
    ]))
    elements.append(t)
    elements.append(Spacer(1, 0.5*cm))

    # Seção 3: Recomendações para Reserva de Emergência
    elements.append(Paragraph("3. Recomendações para Reserva de Emergência", subtitle_style))
    elements.append(Paragraph(f"Baseado nos seus gastos mensais de {formatar_moeda(gastos)}, recomendamos:", normal_style))

    # Tabela de reserva de emergência
    dados_tabela_reserva = [
        ["Tipo de Reserva", "Valor Recomendado"],
        ["Reserva mínima (3x gastos)", formatar_moeda(reserva_3x)],
        ["Reserva ideal (6x gastos)", formatar_moeda(reserva_6x)]
    ]

    t_reserva = Table(dados_tabela_reserva, colWidths=[doc.width/2.0]*2)
    t_reserva.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (1, 0), colors.lightblue),
        ('TEXTCOLOR', (0, 0), (1, 0), colors.white),
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 10),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('GRID', (0, 0), (-1, -1), 1, colors.lightgrey),
    ]))
    elements.append(t_reserva)
    elements.append(Spacer(1, 0.5*cm))

    # Seção 4: Plano Financeiro Personalizado
    elements.append(Paragraph("4. Seu Plano Financeiro Personalizado", subtitle_style))

    # Divide o plano personalizado em parágrafos
    plano_texto = dados_usuario.get('plano_personalizado', '')
    paragrafos = plano_texto.split('\n\n')
    for paragrafo in paragrafos:
        if paragrafo.strip():
            elements.append(Paragraph(paragrafo, normal_style))

    elements.append(Spacer(1, 0.5*cm))

    # Seção 5: Próximos Passos Recomendados
    elements.append(Paragraph("5. Próximos Passos Recomendados", subtitle_style))
    elementos_passos = [
        "Estabeleça um orçamento rigoroso para controlar seus gastos",
        "Priorize a criação da sua reserva de emergência",
        "Reduza ou elimine dívidas de alto custo",
        "Comece a investir assim que tiver sua reserva de emergência completa"
    ]

    for i, passo in enumerate(elementos_passos):
        elements.append(Paragraph(f"{i+1}. {passo}", normal_style))

    # Se houver interesse em investimentos, adicione a seção de estratégias Barsi
    if "sim" in dados_usuario.get('pergunta_5', '').lower():
        elements.append(Spacer(1, 0.5*cm))
        elements.append(Paragraph("6. Estratégias de Investimento (Método Barsi)", subtitle_style))

        estrategias = [
            "<b>Foco em dividendos consistentes</b><br/>- Busque empresas com histórico sólido de pagamento de dividendos<br/>- Prefira empresas que distribuem pelo menos 6% ao ano",
            "<b>Critérios de seleção</b><br/>- Governança corporativa forte<br/>- Liderança no segmento<br/>- Baixo endividamento<br/>- Negócios compreensíveis e estáveis",
            "<b>Estratégia de longo prazo</b><br/>- Invista pensando em décadas<br/>- Reinvista os dividendos para criar o \"efeito bola de neve\"<br/>- Torne-se sócio das empresas",
            "<b>Comece com qualquer valor disponível</b><br/>- Investimentos regulares mensais<br/>- Paciência e disciplina são fundamentais"
        ]

        for i, estrategia in enumerate(estrategias):
            elements.append(Paragraph(f"{i+1}. {estrategia}", normal_style))

    # Cria o PDF
    doc.build(elements)

    print(f"\nRelatório PDF gerado com sucesso: {pdf_filename}")
    return pdf_filename

# Função para gerar relatório em formato markdown (mantida para compatibilidade)
def gerar_relatorio_markdown(dados_usuario):
    hoje = datetime.now().strftime("%d/%m/%Y")

    # Extrai e converte valores financeiros
    salario = converter_para_numero(dados_usuario.get('salario', 0))
    renda_extra = converter_para_numero(dados_usuario.get('renda_extra', 0))
    gastos = converter_para_numero(dados_usuario.get('gastos', 0))
    renda_total = salario + renda_extra

    # Cálculo da reserva de emergência recomendada
    reserva_3x = gastos * 3
    reserva_6x = gastos * 6

    markdown = f"""
# Relatório de Planejamento Financeiro
**Data:** {hoje}

## 1. Perfil Financeiro

**Situação atual:** {dados_usuario.get('pergunta_1', 'Não informado')}
**Estratégias atuais:** {dados_usuario.get('pergunta_2', 'Não informado')}
**Reserva de emergência:** {dados_usuario.get('pergunta_3', 'Não informado')}
**Objetivos financeiros:** {dados_usuario.get('pergunta_4', 'Não informado')}
**Interesse em investimentos:** {dados_usuario.get('pergunta_5', 'Não informado')}

## 2. Análise de Renda e Despesas

**Salário mensal:** R$ {salario:.2f}
**Renda extra mensal:** R$ {renda_extra:.2f}
**Renda total mensal:** R$ {renda_total:.2f}
**Gastos mensais:** R$ {gastos:.2f}
**Saldo mensal:** R$ {renda_total - gastos:.2f}

## 3. Recomendações para Reserva de Emergência

Baseado nos seus gastos mensais de R$ {gastos:.2f}, recomendamos:

- **Reserva mínima (3x gastos):** R$ {reserva_3x:.2f}
- **Reserva ideal (6x gastos):** R$ {reserva_6x:.2f}

## 4. Seu Plano Financeiro Personalizado

{dados_usuario.get('plano_personalizado', '')}

## 5. Próximos Passos Recomendados

1. Estabeleça um orçamento rigoroso para controlar seus gastos
2. Priorize a criação da sua reserva de emergência
3. Reduza ou elimine dívidas de alto custo
4. Comece a investir assim que tiver sua reserva de emergência completa
"""

    # Se houver interesse em investimentos, adicione a seção de estratégias Barsi
    if "sim" in dados_usuario.get('pergunta_5', '').lower():
        markdown += """
## 6. Estratégias de Investimento (Método Barsi)

1. **Foco em dividendos consistentes**
   - Busque empresas com histórico sólido de pagamento de dividendos
   - Prefira empresas que distribuem pelo menos 6% ao ano

2. **Critérios de seleção**
   - Governança corporativa forte
   - Liderança no segmento
   - Baixo endividamento
   - Negócios compreensíveis e estáveis

3. **Estratégia de longo prazo**
   - Invista pensando em décadas
   - Reinvista os dividendos para criar o "efeito bola de neve"
   - Torne-se sócio das empresas

4. **Comece com qualquer valor disponível**
   - Investimentos regulares mensais
   - Paciência e disciplina são fundamentais
"""

    return markdown

# Função principal do Educador Financeiro
def educador_financeiro():
    print("=" * 60)
    print("🌟 BEM-VINDO AO EDUCADOR FINANCEIRO 🌟")
    print("Um programa para ajudar você a melhorar sua saúde financeira")
    print("=" * 60)
    print("Digite 'sair' a qualquer momento para encerrar o programa.\n")

    # Dicionário para armazenar todas as informações do usuário
    dados_usuario = {}

    # Perguntas sobre salário e gastos
    print("\nQual é o seu salário atual? (Informe apenas o valor, ex: 2500)")
    salario_str = input("Valor do salário: R$ ")

    if salario_str.lower() == 'sair':
        print("\nObrigado por usar o Educador Financeiro. Até a próxima!")
        return

    salario = converter_para_numero(salario_str)
    dados_usuario['salario'] = salario

    # Pergunta sobre renda extra
    print("\nVocê possui alguma renda extra mensal? (sim/não)")
    tem_renda_extra = input("Resposta: ")

    if tem_renda_extra.lower() == 'sair':
        print("\nObrigado por usar o Educador Financeiro. Até a próxima!")
        return

    renda_extra = 0
    if tem_renda_extra.lower() in ['sim', 's', 'yes', 'y']:
        print("\nQual é o valor da sua renda extra mensal?")
        renda_extra_str = input("Valor da renda extra: R$ ")

        if renda_extra_str.lower() == 'sair':
            print("\nObrigado por usar o Educador Financeiro. Até a próxima!")
            return

        renda_extra = converter_para_numero(renda_extra_str)

    dados_usuario['renda_extra'] = renda_extra
    renda_total = salario + renda_extra
    dados_usuario['renda_total'] = renda_total

    # Pergunta sobre gastos mensais
    print("\nQual é o valor total dos seus gastos mensais? (Considerando todas as contas, despesas e dívidas)")
    gastos_str = input("Valor dos gastos: R$ ")

    if gastos_str.lower() == 'sair':
        print("\nObrigado por usar o Educador Financeiro. Até a próxima!")
        return

    gastos = converter_para_numero(gastos_str)
    dados_usuario['gastos'] = gastos

    # Cálculo e exibição da reserva de emergência necessária
    reserva_3x = gastos * 3
    reserva_6x = gastos * 6

    print(f"\n💰 Com base nos seus gastos mensais de R$ {gastos:.2f}:")
    print(f"   - Sua reserva de emergência mínima (3x) deveria ser de R$ {reserva_3x:.2f}")
    print(f"   - Sua reserva de emergência ideal (6x) deveria ser de R$ {reserva_6x:.2f}")

    if renda_total < gastos:
        diferenca = gastos - renda_total
        print(f"\n⚠️ Atenção! Seus gastos excedem sua renda em R$ {diferenca:.2f} por mês.")
        print("   Você precisa reduzir despesas ou buscar uma renda extra neste valor para equilibrar suas finanças.")
    else:
        sobra = renda_total - gastos
        print(f"\n✅ Você possui um saldo positivo de R$ {sobra:.2f} por mês.")
        print("   Este valor pode ser direcionado para sua reserva de emergência ou investimentos.")

    # Perguntas iniciais
    perguntas = [
        "Para começarmos, poderia me contar qual é a sua atual situação financeira? (Por exemplo: estou endividado, conseguindo pagar as contas mas sem sobras, consigo poupar um pouco, etc.)",
        "O que você está fazendo atualmente para lidar com suas dívidas ou melhorar sua situação financeira?",
        "Você possui alguma reserva de emergência? Se sim, por quanto tempo ela conseguiria cobrir suas despesas básicas?",
        "Quais são seus principais objetivos financeiros para os próximos meses ou anos?",
        "Você tem interesse em aprender sobre investimentos em ações e dividendos?"
    ]

    # Coleta respostas para as perguntas iniciais
    for i, pergunta in enumerate(perguntas):
        print(f"\n{pergunta}")
        resposta = input("Sua resposta: ")

        if resposta.lower() == 'sair':
            print("\nObrigado por usar o Educador Financeiro. Até a próxima!")
            return

        dados_usuario[f"pergunta_{i+1}"] = resposta

        # Após coletar a resposta para cada pergunta, fornece feedback personalizado
        # Para a última pergunta sobre investimentos, incluímos as estratégias de Barsi se houver interesse
        incluir_barsi = False
        if i == 4:  # Última pergunta (sobre investimentos)
            incluir_barsi = "sim" in resposta.lower() or "interesse" in resposta.lower()

        prompt_feedback = f"""
        O usuário respondeu à pergunta: "{pergunta}"
        com a resposta: "{resposta}".

        Forneça um feedback acolhedor, sem julgamentos e com orientações personalizadas
        com base na resposta dele. Seja breve (2-3 parágrafos no máximo).
        """

        feedback = obter_resposta(prompt_feedback, incluir_barsi)
        print(f"\n🌿 Análise & Orientação 🌿\n{feedback}\n")

    # Análise final e recomendações
    prompt_final = f"""
    Com base nas respostas do usuário:

    Salário mensal: R$ {salario:.2f}
    Renda extra mensal: R$ {renda_extra:.2f}
    Renda total mensal: R$ {renda_total:.2f}
    Gastos mensais: R$ {gastos:.2f}
    Situação financeira: {dados_usuario.get('pergunta_1', 'Não informado')}
    Estratégias atuais: {dados_usuario.get('pergunta_2', 'Não informado')}
    Reserva de emergência: {dados_usuario.get('pergunta_3', 'Não informado')}
    Objetivos financeiros: {dados_usuario.get('pergunta_4', 'Não informado')}
    Interesse em investimentos: {dados_usuario.get('pergunta_5', 'Não informado')}

    Forneça uma análise completa e um plano personalizado de educação financeira.

    IMPORTANTE:
    1. Enfatize a importância de ter uma reserva de emergência de 3 a 6 vezes o valor total das despesas mensais (R$ {reserva_3x:.2f} a R$ {reserva_6x:.2f}).
    2. Comente sobre o equilíbrio entre renda e despesas.
    3. Se o usuário demonstrou interesse em investimentos, inclua algumas estratégias de Luiz Barsi sobre investimentos em dividendos.

    Apresente o plano de forma estruturada com passos claros e práticos.
    """

    # Decide se inclui as estratégias de Barsi com base no interesse do usuário
    incluir_barsi = "sim" in dados_usuario.get('pergunta_5', '').lower() or "interesse" in dados_usuario.get('pergunta_5', '').lower()

    plano_personalizado = obter_resposta(prompt_final, incluir_barsi)
    dados_usuario['plano_personalizado'] = plano_personalizado

    print("\n" + "=" * 60)
    print("📊 SEU PLANO FINANCEIRO PERSONALIZADO 📊")
    print("=" * 60 + "\n")
    print(plano_personalizado)
    print("\n" + "=" * 60)

    # Perguntas adicionais
    continuar = True
    while continuar:
        print("\nVocê tem alguma dúvida específica sobre finanças ou gostaria de mais detalhes sobre algum aspecto?")
        print("Sugestões: 'estratégias de Barsi', 'dividendos', 'como começar a investir', 'como sair das dívidas'")
        print("Digite 'sair' para encerrar, 'relatório' para gerar seu relatório financeiro em texto, ou 'pdf' para gerar em PDF.")

        pergunta_adicional = input("Sua pergunta: ")

        if pergunta_adicional.lower() == 'sair':
            print("\nDeseja gerar um relatório com suas informações financeiras? (texto/pdf/não)")
            gerar_relatorio = input("Resposta: ")

            if gerar_relatorio.lower() in ['texto', 't', 'text']:
                relatorio = gerar_relatorio_markdown(dados_usuario)
                print("\n" + "=" * 60)
                print("📑 SEU RELATÓRIO FINANCEIRO COMPLETO 📑")
                print("=" * 60 + "\n")
                print(relatorio)
                print("\n" + "=" * 60)
                print("\nRelatório em texto gerado com sucesso!")
            elif gerar_relatorio.lower() in ['pdf', 'p']:
                pdf_filename = gerar_relatorio_pdf(dados_usuario)
                print(f"\nRelatório PDF gerado com sucesso: {pdf_filename}")

            print("\nObrigado por usar o Educador Financeiro. Desejo sucesso na sua jornada financeira!")
            continuar = False
            continue

        if pergunta_adicional.lower() == 'relatório':
            relatorio = gerar_relatorio_markdown(dados_usuario)
            print("\n" + "=" * 60)
            print("📑 SEU RELATÓRIO FINANCEIRO COMPLETO 📑")
            print("=" * 60 + "\n")
            print(relatorio)
            print("\n" + "=" * 60)
            continue

        if pergunta_adicional.lower() == 'pdf':
            pdf_filename = gerar_relatorio_pdf(dados_usuario)
            continue

        # Comando especial para ver diretamente as estratégias de Barsi
        if pergunta_adicional.lower() == 'estratégias de barsi' or pergunta_adicional.lower() == 'barsi':
            print("\n📈 ESTRATÉGIAS DE INVESTIMENTO DE LUIZ BARSI 📈\n")
            estrategias = estrategias_barsi.replace("Estratégias de Investimento de Luiz Barsi:", "")
            print(estrategias)
            continue

        # Verifica se a pergunta é sobre Barsi ou dividendos
        incluir_barsi = any(termo in pergunta_adicional.lower() for termo in ["barsi", "divid", "investimento", "ações"])

        resposta_adicional = obter_resposta(pergunta_adicional, incluir_barsi)
        print(f"\n💡 Resposta:\n{resposta_adicional}\n")

# Executa o programa
if __name__ == "__main__":
    educador_financeiro()
```
