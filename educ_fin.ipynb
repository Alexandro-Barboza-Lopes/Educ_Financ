```python
import os
from google import genai
from google.generativeai import types
from IPython import get_ipython
from IPython.display import display
import os
from google.colab import userdata
import google.generativeai as genai
from datetime import datetime
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib.units import inch, cm
from reportlab.lib.enums import TA_CENTER, TA_JUSTIFY
import tempfile

# ConfiguraÃ§Ã£o da API - substitua pela sua chave
# VocÃª pode usar os.environ['GOOGLE_API_KEY'] = "sua_chave_aqui" diretamente no cÃ³digo
# Ou usar a abordagem do userdata do Google Colab como no exemplo original

userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))
# Inicializa o modelo e a sessÃ£o de chat
model = genai.GenerativeModel('gemini-2.0-flash')

# DefiniÃ§Ã£o das instruÃ§Ãµes para o comportamento do assistente
instrucoes_sistema = """
VocÃª Ã© um educador financeiro experiente, calmo e acolhedor com mais de 20 anos de experiÃªncia.
Seu papel Ã©:
- Fornecer orientaÃ§Ã£o financeira personalizada com base nas respostas do usuÃ¡rio
- Ser atencioso, empÃ¡tico e nunca julgar a situaÃ§Ã£o financeira do usuÃ¡rio
- Oferecer conselhos prÃ¡ticos e realistas considerando a situaÃ§Ã£o especÃ­fica
- Enfatizar a importÃ¢ncia de ter uma reserva de emergÃªncia de 3 a 6 vezes o valor total das despesas mensais
- Sugerir estratÃ©gias para gerenciar dÃ­vidas e melhorar a saÃºde financeira
- Manter um tom calmo, positivo e encorajador em todas as interaÃ§Ãµes
- Compartilhar, quando apropriado, as estratÃ©gias de investimento de Luiz Barsi, especialmente sobre dividendos
"""

# EstratÃ©gias de investimento de Luiz Barsi - para uso nas orientaÃ§Ãµes
estrategias_barsi = """
EstratÃ©gias de Investimento de Luiz Barsi:

1. Foco em empresas pagadoras de bons dividendos:
   - Buscar empresas sÃ³lidas com histÃ³rico de pagamento consistente de dividendos
   - Preferir empresas que distribuem regularmente pelo menos 6% ao ano em dividendos
   - Reinvestir os dividendos para criar o "efeito bola de neve"

2. CritÃ©rios de seleÃ§Ã£o de empresas segundo Barsi:
   - Empresas com forte governanÃ§a corporativa
   - LÃ­deres em seus segmentos
   - Baixo endividamento
   - NegÃ³cios simples de entender e pouco afetados por inovaÃ§Ãµes tecnolÃ³gicas
   - HistÃ³rico de resultados crescentes ou estÃ¡veis
   - Que atuam em setores essenciais da economia

3. Filosofia de longo prazo:
   - Investir pensando em dÃ©cadas, nÃ£o em dias ou meses
   - NÃ£o se preocupar com oscilaÃ§Ãµes de curto prazo do mercado
   - Buscar uma renda passiva crescente atravÃ©s dos dividendos
   - Tornar-se "sÃ³cio" das empresas e nÃ£o apenas especulador

4. GestÃ£o da carteira:
   - Diversificar em 8-12 empresas de qualidade
   - Acompanhar os resultados trimestrais das empresas
   - Aumentar posiÃ§Ã£o em empresas que mantÃªm bons fundamentos
   - Avaliar constantemente se os motivos que o levaram a investir na empresa continuam vÃ¡lidos

5. RecomendaÃ§Ãµes prÃ¡ticas:
   - ComeÃ§ar com qualquer valor, mesmo que pequeno
   - Investir com regularidade (mensalmente)
   - Reinvestir os dividendos para acelerar o crescimento do patrimÃ´nio
   - Ter paciÃªncia e disciplina para colher os resultados no longo prazo
"""

# Cria um objeto de chat com as instruÃ§Ãµes do sistema
chat = model.start_chat(history=[])

# FunÃ§Ã£o para obter respostas do modelo AI
def obter_resposta(mensagem, incluir_barsi=False):
    # Adiciona as instruÃ§Ãµes ao sistema para cada mensagem
    if incluir_barsi:
        resposta = chat.send_message(f"{instrucoes_sistema}\n\n{estrategias_barsi}\n\nUsuÃ¡rio: {mensagem}")
    else:
        resposta = chat.send_message(f"{instrucoes_sistema}\n\nUsuÃ¡rio: {mensagem}")
    return resposta.text

# FunÃ§Ã£o para converter string em valor numÃ©rico, removendo caracteres nÃ£o numÃ©ricos
def converter_para_numero(valor_str):
    # Remove R$, pontos, vÃ­rgulas e espaÃ§os, e substitui vÃ­rgula por ponto
    if isinstance(valor_str, str):
        valor_limpo = valor_str.replace("R$", "").replace(".", "").replace(",", ".").strip()
        # Tenta converter para float
        try:
            return float(valor_limpo)
        except ValueError:
            return 0
    elif isinstance(valor_str, (int, float)):
        return float(valor_str)
    else:
        return 0

# FunÃ§Ã£o para formatar valores monetÃ¡rios
def formatar_moeda(valor):
    return f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

# FunÃ§Ã£o para gerar relatÃ³rio em PDF
def gerar_relatorio_pdf(dados_usuario):
    # Cria um arquivo temporÃ¡rio para o PDF
    pdf_filename = f"relatorio_financeiro_{datetime.now().strftime('%Y%m%d_%H%M%S')}.pdf"

    # Extrai e converte valores financeiros
    salario = converter_para_numero(dados_usuario.get('salario', 0))
    renda_extra = converter_para_numero(dados_usuario.get('renda_extra', 0))
    gastos = converter_para_numero(dados_usuario.get('gastos', 0))
    renda_total = salario + renda_extra

    # CÃ¡lculo da reserva de emergÃªncia recomendada
    reserva_3x = gastos * 3
    reserva_6x = gastos * 6

    # Estilos para o PDF
    styles = getSampleStyleSheet()

    # Estilo para o tÃ­tulo principal
    title_style = ParagraphStyle(
        'TitleStyle',
        parent=styles['Heading1'],
        fontSize=18,
        alignment=TA_CENTER,
        spaceAfter=16,
        textColor=colors.darkblue,
    )

    # Estilo para os subtÃ­tulos
    subtitle_style = ParagraphStyle(
        'SubtitleStyle',
        parent=styles['Heading2'],
        fontSize=14,
        spaceAfter=10,
        textColor=colors.darkblue,
    )

    # Estilo para texto normal
    normal_style = ParagraphStyle(
        'NormalStyle',
        parent=styles['Normal'],
        fontSize=10,
        alignment=TA_JUSTIFY,
        spaceAfter=8,
    )

    # Estilo para informaÃ§Ãµes financeiras
    info_style = ParagraphStyle(
        'InfoStyle',
        parent=styles['Normal'],
        fontSize=10,
        spaceAfter=6,
    )

    # Cria o documento PDF
    doc = SimpleDocTemplate(
        pdf_filename,
        pagesize=A4,
        rightMargin=72,
        leftMargin=72,
        topMargin=72,
        bottomMargin=72
    )

    # Lista de elementos para o PDF
    elements = []

    # Data do relatÃ³rio
    data_atual = datetime.now().strftime("%d/%m/%Y")

    # Adiciona o tÃ­tulo
    elements.append(Paragraph("RelatÃ³rio de Planejamento Financeiro", title_style))
    elements.append(Paragraph(f"Data: {data_atual}", info_style))
    elements.append(Spacer(1, 0.5*cm))

    # SeÃ§Ã£o 1: Perfil Financeiro
    elements.append(Paragraph("1. Perfil Financeiro", subtitle_style))
    elements.append(Paragraph(f"<b>SituaÃ§Ã£o atual:</b> {dados_usuario.get('pergunta_1', 'NÃ£o informado')}", normal_style))
    elements.append(Paragraph(f"<b>EstratÃ©gias atuais:</b> {dados_usuario.get('pergunta_2', 'NÃ£o informado')}", normal_style))
    elements.append(Paragraph(f"<b>Reserva de emergÃªncia:</b> {dados_usuario.get('pergunta_3', 'NÃ£o informado')}", normal_style))
    elements.append(Paragraph(f"<b>Objetivos financeiros:</b> {dados_usuario.get('pergunta_4', 'NÃ£o informado')}", normal_style))
    elements.append(Paragraph(f"<b>Interesse em investimentos:</b> {dados_usuario.get('pergunta_5', 'NÃ£o informado')}", normal_style))
    elements.append(Spacer(1, 0.5*cm))

    # SeÃ§Ã£o 2: AnÃ¡lise de Renda e Despesas
    elements.append(Paragraph("2. AnÃ¡lise de Renda e Despesas", subtitle_style))

    # Tabela de renda e despesas
    dados_tabela = [
        ["Item", "Valor"],
        ["SalÃ¡rio mensal", formatar_moeda(salario)],
        ["Renda extra mensal", formatar_moeda(renda_extra)],
        ["Renda total mensal", formatar_moeda(renda_total)],
        ["Gastos mensais", formatar_moeda(gastos)],
        ["Saldo mensal", formatar_moeda(renda_total - gastos)]
    ]

    t = Table(dados_tabela, colWidths=[doc.width/2.0]*2)
    t.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (1, 0), colors.lightblue),
        ('TEXTCOLOR', (0, 0), (1, 0), colors.white),
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 10),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('GRID', (0, 0), (-1, -1), 1, colors.lightgrey),
    ]))
    elements.append(t)
    elements.append(Spacer(1, 0.5*cm))

    # SeÃ§Ã£o 3: RecomendaÃ§Ãµes para Reserva de EmergÃªncia
    elements.append(Paragraph("3. RecomendaÃ§Ãµes para Reserva de EmergÃªncia", subtitle_style))
    elements.append(Paragraph(f"Baseado nos seus gastos mensais de {formatar_moeda(gastos)}, recomendamos:", normal_style))

    # Tabela de reserva de emergÃªncia
    dados_tabela_reserva = [
        ["Tipo de Reserva", "Valor Recomendado"],
        ["Reserva mÃ­nima (3x gastos)", formatar_moeda(reserva_3x)],
        ["Reserva ideal (6x gastos)", formatar_moeda(reserva_6x)]
    ]

    t_reserva = Table(dados_tabela_reserva, colWidths=[doc.width/2.0]*2)
    t_reserva.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (1, 0), colors.lightblue),
        ('TEXTCOLOR', (0, 0), (1, 0), colors.white),
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 10),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('GRID', (0, 0), (-1, -1), 1, colors.lightgrey),
    ]))
    elements.append(t_reserva)
    elements.append(Spacer(1, 0.5*cm))

    # SeÃ§Ã£o 4: Plano Financeiro Personalizado
    elements.append(Paragraph("4. Seu Plano Financeiro Personalizado", subtitle_style))

    # Divide o plano personalizado em parÃ¡grafos
    plano_texto = dados_usuario.get('plano_personalizado', '')
    paragrafos = plano_texto.split('\n\n')
    for paragrafo in paragrafos:
        if paragrafo.strip():
            elements.append(Paragraph(paragrafo, normal_style))

    elements.append(Spacer(1, 0.5*cm))

    # SeÃ§Ã£o 5: PrÃ³ximos Passos Recomendados
    elements.append(Paragraph("5. PrÃ³ximos Passos Recomendados", subtitle_style))
    elementos_passos = [
        "EstabeleÃ§a um orÃ§amento rigoroso para controlar seus gastos",
        "Priorize a criaÃ§Ã£o da sua reserva de emergÃªncia",
        "Reduza ou elimine dÃ­vidas de alto custo",
        "Comece a investir assim que tiver sua reserva de emergÃªncia completa"
    ]

    for i, passo in enumerate(elementos_passos):
        elements.append(Paragraph(f"{i+1}. {passo}", normal_style))

    # Se houver interesse em investimentos, adicione a seÃ§Ã£o de estratÃ©gias Barsi
    if "sim" in dados_usuario.get('pergunta_5', '').lower():
        elements.append(Spacer(1, 0.5*cm))
        elements.append(Paragraph("6. EstratÃ©gias de Investimento (MÃ©todo Barsi)", subtitle_style))

        estrategias = [
            "<b>Foco em dividendos consistentes</b><br/>- Busque empresas com histÃ³rico sÃ³lido de pagamento de dividendos<br/>- Prefira empresas que distribuem pelo menos 6% ao ano",
            "<b>CritÃ©rios de seleÃ§Ã£o</b><br/>- GovernanÃ§a corporativa forte<br/>- LideranÃ§a no segmento<br/>- Baixo endividamento<br/>- NegÃ³cios compreensÃ­veis e estÃ¡veis",
            "<b>EstratÃ©gia de longo prazo</b><br/>- Invista pensando em dÃ©cadas<br/>- Reinvista os dividendos para criar o \"efeito bola de neve\"<br/>- Torne-se sÃ³cio das empresas",
            "<b>Comece com qualquer valor disponÃ­vel</b><br/>- Investimentos regulares mensais<br/>- PaciÃªncia e disciplina sÃ£o fundamentais"
        ]

        for i, estrategia in enumerate(estrategias):
            elements.append(Paragraph(f"{i+1}. {estrategia}", normal_style))

    # Cria o PDF
    doc.build(elements)

    print(f"\nRelatÃ³rio PDF gerado com sucesso: {pdf_filename}")
    return pdf_filename

# FunÃ§Ã£o para gerar relatÃ³rio em formato markdown (mantida para compatibilidade)
def gerar_relatorio_markdown(dados_usuario):
    hoje = datetime.now().strftime("%d/%m/%Y")

    # Extrai e converte valores financeiros
    salario = converter_para_numero(dados_usuario.get('salario', 0))
    renda_extra = converter_para_numero(dados_usuario.get('renda_extra', 0))
    gastos = converter_para_numero(dados_usuario.get('gastos', 0))
    renda_total = salario + renda_extra

    # CÃ¡lculo da reserva de emergÃªncia recomendada
    reserva_3x = gastos * 3
    reserva_6x = gastos * 6

    markdown = f"""
# RelatÃ³rio de Planejamento Financeiro
**Data:** {hoje}

## 1. Perfil Financeiro

**SituaÃ§Ã£o atual:** {dados_usuario.get('pergunta_1', 'NÃ£o informado')}
**EstratÃ©gias atuais:** {dados_usuario.get('pergunta_2', 'NÃ£o informado')}
**Reserva de emergÃªncia:** {dados_usuario.get('pergunta_3', 'NÃ£o informado')}
**Objetivos financeiros:** {dados_usuario.get('pergunta_4', 'NÃ£o informado')}
**Interesse em investimentos:** {dados_usuario.get('pergunta_5', 'NÃ£o informado')}

## 2. AnÃ¡lise de Renda e Despesas

**SalÃ¡rio mensal:** R$ {salario:.2f}
**Renda extra mensal:** R$ {renda_extra:.2f}
**Renda total mensal:** R$ {renda_total:.2f}
**Gastos mensais:** R$ {gastos:.2f}
**Saldo mensal:** R$ {renda_total - gastos:.2f}

## 3. RecomendaÃ§Ãµes para Reserva de EmergÃªncia

Baseado nos seus gastos mensais de R$ {gastos:.2f}, recomendamos:

- **Reserva mÃ­nima (3x gastos):** R$ {reserva_3x:.2f}
- **Reserva ideal (6x gastos):** R$ {reserva_6x:.2f}

## 4. Seu Plano Financeiro Personalizado

{dados_usuario.get('plano_personalizado', '')}

## 5. PrÃ³ximos Passos Recomendados

1. EstabeleÃ§a um orÃ§amento rigoroso para controlar seus gastos
2. Priorize a criaÃ§Ã£o da sua reserva de emergÃªncia
3. Reduza ou elimine dÃ­vidas de alto custo
4. Comece a investir assim que tiver sua reserva de emergÃªncia completa
"""

    # Se houver interesse em investimentos, adicione a seÃ§Ã£o de estratÃ©gias Barsi
    if "sim" in dados_usuario.get('pergunta_5', '').lower():
        markdown += """
## 6. EstratÃ©gias de Investimento (MÃ©todo Barsi)

1. **Foco em dividendos consistentes**
   - Busque empresas com histÃ³rico sÃ³lido de pagamento de dividendos
   - Prefira empresas que distribuem pelo menos 6% ao ano

2. **CritÃ©rios de seleÃ§Ã£o**
   - GovernanÃ§a corporativa forte
   - LideranÃ§a no segmento
   - Baixo endividamento
   - NegÃ³cios compreensÃ­veis e estÃ¡veis

3. **EstratÃ©gia de longo prazo**
   - Invista pensando em dÃ©cadas
   - Reinvista os dividendos para criar o "efeito bola de neve"
   - Torne-se sÃ³cio das empresas

4. **Comece com qualquer valor disponÃ­vel**
   - Investimentos regulares mensais
   - PaciÃªncia e disciplina sÃ£o fundamentais
"""

    return markdown

# FunÃ§Ã£o principal do Educador Financeiro
def educador_financeiro():
    print("=" * 60)
    print("ğŸŒŸ BEM-VINDO AO EDUCADOR FINANCEIRO ğŸŒŸ")
    print("Um programa para ajudar vocÃª a melhorar sua saÃºde financeira")
    print("=" * 60)
    print("Digite 'sair' a qualquer momento para encerrar o programa.\n")

    # DicionÃ¡rio para armazenar todas as informaÃ§Ãµes do usuÃ¡rio
    dados_usuario = {}

    # Perguntas sobre salÃ¡rio e gastos
    print("\nQual Ã© o seu salÃ¡rio atual? (Informe apenas o valor, ex: 2500)")
    salario_str = input("Valor do salÃ¡rio: R$ ")

    if salario_str.lower() == 'sair':
        print("\nObrigado por usar o Educador Financeiro. AtÃ© a prÃ³xima!")
        return

    salario = converter_para_numero(salario_str)
    dados_usuario['salario'] = salario

    # Pergunta sobre renda extra
    print("\nVocÃª possui alguma renda extra mensal? (sim/nÃ£o)")
    tem_renda_extra = input("Resposta: ")

    if tem_renda_extra.lower() == 'sair':
        print("\nObrigado por usar o Educador Financeiro. AtÃ© a prÃ³xima!")
        return

    renda_extra = 0
    if tem_renda_extra.lower() in ['sim', 's', 'yes', 'y']:
        print("\nQual Ã© o valor da sua renda extra mensal?")
        renda_extra_str = input("Valor da renda extra: R$ ")

        if renda_extra_str.lower() == 'sair':
            print("\nObrigado por usar o Educador Financeiro. AtÃ© a prÃ³xima!")
            return

        renda_extra = converter_para_numero(renda_extra_str)

    dados_usuario['renda_extra'] = renda_extra
    renda_total = salario + renda_extra
    dados_usuario['renda_total'] = renda_total

    # Pergunta sobre gastos mensais
    print("\nQual Ã© o valor total dos seus gastos mensais? (Considerando todas as contas, despesas e dÃ­vidas)")
    gastos_str = input("Valor dos gastos: R$ ")

    if gastos_str.lower() == 'sair':
        print("\nObrigado por usar o Educador Financeiro. AtÃ© a prÃ³xima!")
        return

    gastos = converter_para_numero(gastos_str)
    dados_usuario['gastos'] = gastos

    # CÃ¡lculo e exibiÃ§Ã£o da reserva de emergÃªncia necessÃ¡ria
    reserva_3x = gastos * 3
    reserva_6x = gastos * 6

    print(f"\nğŸ’° Com base nos seus gastos mensais de R$ {gastos:.2f}:")
    print(f"   - Sua reserva de emergÃªncia mÃ­nima (3x) deveria ser de R$ {reserva_3x:.2f}")
    print(f"   - Sua reserva de emergÃªncia ideal (6x) deveria ser de R$ {reserva_6x:.2f}")

    if renda_total < gastos:
        diferenca = gastos - renda_total
        print(f"\nâš ï¸ AtenÃ§Ã£o! Seus gastos excedem sua renda em R$ {diferenca:.2f} por mÃªs.")
        print("   VocÃª precisa reduzir despesas ou buscar uma renda extra neste valor para equilibrar suas finanÃ§as.")
    else:
        sobra = renda_total - gastos
        print(f"\nâœ… VocÃª possui um saldo positivo de R$ {sobra:.2f} por mÃªs.")
        print("   Este valor pode ser direcionado para sua reserva de emergÃªncia ou investimentos.")

    # Perguntas iniciais
    perguntas = [
        "Para comeÃ§armos, poderia me contar qual Ã© a sua atual situaÃ§Ã£o financeira? (Por exemplo: estou endividado, conseguindo pagar as contas mas sem sobras, consigo poupar um pouco, etc.)",
        "O que vocÃª estÃ¡ fazendo atualmente para lidar com suas dÃ­vidas ou melhorar sua situaÃ§Ã£o financeira?",
        "VocÃª possui alguma reserva de emergÃªncia? Se sim, por quanto tempo ela conseguiria cobrir suas despesas bÃ¡sicas?",
        "Quais sÃ£o seus principais objetivos financeiros para os prÃ³ximos meses ou anos?",
        "VocÃª tem interesse em aprender sobre investimentos em aÃ§Ãµes e dividendos?"
    ]

    # Coleta respostas para as perguntas iniciais
    for i, pergunta in enumerate(perguntas):
        print(f"\n{pergunta}")
        resposta = input("Sua resposta: ")

        if resposta.lower() == 'sair':
            print("\nObrigado por usar o Educador Financeiro. AtÃ© a prÃ³xima!")
            return

        dados_usuario[f"pergunta_{i+1}"] = resposta

        # ApÃ³s coletar a resposta para cada pergunta, fornece feedback personalizado
        # Para a Ãºltima pergunta sobre investimentos, incluÃ­mos as estratÃ©gias de Barsi se houver interesse
        incluir_barsi = False
        if i == 4:  # Ãšltima pergunta (sobre investimentos)
            incluir_barsi = "sim" in resposta.lower() or "interesse" in resposta.lower()

        prompt_feedback = f"""
        O usuÃ¡rio respondeu Ã  pergunta: "{pergunta}"
        com a resposta: "{resposta}".

        ForneÃ§a um feedback acolhedor, sem julgamentos e com orientaÃ§Ãµes personalizadas
        com base na resposta dele. Seja breve (2-3 parÃ¡grafos no mÃ¡ximo).
        """

        feedback = obter_resposta(prompt_feedback, incluir_barsi)
        print(f"\nğŸŒ¿ AnÃ¡lise & OrientaÃ§Ã£o ğŸŒ¿\n{feedback}\n")

    # AnÃ¡lise final e recomendaÃ§Ãµes
    prompt_final = f"""
    Com base nas respostas do usuÃ¡rio:

    SalÃ¡rio mensal: R$ {salario:.2f}
    Renda extra mensal: R$ {renda_extra:.2f}
    Renda total mensal: R$ {renda_total:.2f}
    Gastos mensais: R$ {gastos:.2f}
    SituaÃ§Ã£o financeira: {dados_usuario.get('pergunta_1', 'NÃ£o informado')}
    EstratÃ©gias atuais: {dados_usuario.get('pergunta_2', 'NÃ£o informado')}
    Reserva de emergÃªncia: {dados_usuario.get('pergunta_3', 'NÃ£o informado')}
    Objetivos financeiros: {dados_usuario.get('pergunta_4', 'NÃ£o informado')}
    Interesse em investimentos: {dados_usuario.get('pergunta_5', 'NÃ£o informado')}

    ForneÃ§a uma anÃ¡lise completa e um plano personalizado de educaÃ§Ã£o financeira.

    IMPORTANTE:
    1. Enfatize a importÃ¢ncia de ter uma reserva de emergÃªncia de 3 a 6 vezes o valor total das despesas mensais (R$ {reserva_3x:.2f} a R$ {reserva_6x:.2f}).
    2. Comente sobre o equilÃ­brio entre renda e despesas.
    3. Se o usuÃ¡rio demonstrou interesse em investimentos, inclua algumas estratÃ©gias de Luiz Barsi sobre investimentos em dividendos.

    Apresente o plano de forma estruturada com passos claros e prÃ¡ticos.
    """

    # Decide se inclui as estratÃ©gias de Barsi com base no interesse do usuÃ¡rio
    incluir_barsi = "sim" in dados_usuario.get('pergunta_5', '').lower() or "interesse" in dados_usuario.get('pergunta_5', '').lower()

    plano_personalizado = obter_resposta(prompt_final, incluir_barsi)
    dados_usuario['plano_personalizado'] = plano_personalizado

    print("\n" + "=" * 60)
    print("ğŸ“Š SEU PLANO FINANCEIRO PERSONALIZADO ğŸ“Š")
    print("=" * 60 + "\n")
    print(plano_personalizado)
    print("\n" + "=" * 60)

    # Perguntas adicionais
    continuar = True
    while continuar:
        print("\nVocÃª tem alguma dÃºvida especÃ­fica sobre finanÃ§as ou gostaria de mais detalhes sobre algum aspecto?")
        print("SugestÃµes: 'estratÃ©gias de Barsi', 'dividendos', 'como comeÃ§ar a investir', 'como sair das dÃ­vidas'")
        print("Digite 'sair' para encerrar, 'relatÃ³rio' para gerar seu relatÃ³rio financeiro em texto, ou 'pdf' para gerar em PDF.")

        pergunta_adicional = input("Sua pergunta: ")

        if pergunta_adicional.lower() == 'sair':
            print("\nDeseja gerar um relatÃ³rio com suas informaÃ§Ãµes financeiras? (texto/pdf/nÃ£o)")
            gerar_relatorio = input("Resposta: ")

            if gerar_relatorio.lower() in ['texto', 't', 'text']:
                relatorio = gerar_relatorio_markdown(dados_usuario)
                print("\n" + "=" * 60)
                print("ğŸ“‘ SEU RELATÃ“RIO FINANCEIRO COMPLETO ğŸ“‘")
                print("=" * 60 + "\n")
                print(relatorio)
                print("\n" + "=" * 60)
                print("\nRelatÃ³rio em texto gerado com sucesso!")
            elif gerar_relatorio.lower() in ['pdf', 'p']:
                pdf_filename = gerar_relatorio_pdf(dados_usuario)
                print(f"\nRelatÃ³rio PDF gerado com sucesso: {pdf_filename}")

            print("\nObrigado por usar o Educador Financeiro. Desejo sucesso na sua jornada financeira!")
            continuar = False
            continue

        if pergunta_adicional.lower() == 'relatÃ³rio':
            relatorio = gerar_relatorio_markdown(dados_usuario)
            print("\n" + "=" * 60)
            print("ğŸ“‘ SEU RELATÃ“RIO FINANCEIRO COMPLETO ğŸ“‘")
            print("=" * 60 + "\n")
            print(relatorio)
            print("\n" + "=" * 60)
            continue

        if pergunta_adicional.lower() == 'pdf':
            pdf_filename = gerar_relatorio_pdf(dados_usuario)
            continue

        # Comando especial para ver diretamente as estratÃ©gias de Barsi
        if pergunta_adicional.lower() == 'estratÃ©gias de barsi' or pergunta_adicional.lower() == 'barsi':
            print("\nğŸ“ˆ ESTRATÃ‰GIAS DE INVESTIMENTO DE LUIZ BARSI ğŸ“ˆ\n")
            estrategias = estrategias_barsi.replace("EstratÃ©gias de Investimento de Luiz Barsi:", "")
            print(estrategias)
            continue

        # Verifica se a pergunta Ã© sobre Barsi ou dividendos
        incluir_barsi = any(termo in pergunta_adicional.lower() for termo in ["barsi", "divid", "investimento", "aÃ§Ãµes"])

        resposta_adicional = obter_resposta(pergunta_adicional, incluir_barsi)
        print(f"\nğŸ’¡ Resposta:\n{resposta_adicional}\n")

# Executa o programa
if __name__ == "__main__":
    educador_financeiro()
```
